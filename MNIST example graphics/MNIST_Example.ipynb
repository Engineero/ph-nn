```python
from tensorflow import keras
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Render examples for MNIST digits (test dataset)
good = [1, 6, 5, 7, 9, 35, 13, 8144, 17, 22]
bad  = [7325, 7898, 8883, 1121, 9442, 6848, 494, 15, 418, 3510]
ugly = [5194, 7074, 7732, 9180, 1004, 9450, 9534, 8268, 7851, 3696]
```

```python
(x_train, _), (_, _) = keras.datasets.mnist.load_data()

def plot_digit(axs, offset):
    # axs.axis('off')
    axs.grid(False, visible=False)
    axs.set_frame_on(True)
    axs.get_xaxis().set_ticks([])
    axs.get_yaxis().set_ticks([])
    
    imsample = np.resize(x_train[offset], (28, 28))
    axs.imshow(imsample, cmap='Greys', interpolation='none')
```

```python
fig, axs = plt.subplots(3, 10, sharex=True, sharey=True, figsize=(10,3))

for i in range(10):
    plot_digit(axs[0, i], good[i])
    plot_digit(axs[1, i], bad[i])
    plot_digit(axs[2, i], ugly[i])

plt.savefig('mnist_example.pdf', bbox_inches='tight')
```
